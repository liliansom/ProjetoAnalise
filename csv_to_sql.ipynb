```python
# Importação das bibliotecas
import pandas as pd
import matplotlib as plt
import pyodbc
```

```python
# Criação do Dataframe
file = pd.read_csv(r'C:\Users\Usuário\Documents\GitHub\ProjetoAnalise\gender_submission.csv', sep=',')
df = pd.DataFrame(file)
df
```

```python
# conexão com BD
dadosconex = "Driver={SQL Server};Server=.;Database=Titanic;"
conex = pyodbc.connect(dadosconex)
cursor = conex.cursor()
```

```python
# Criação da tabela Sobreviventes
cursor.execute(f'''DROP TABLE Sobreviventes''')
# TABELA GENDER_SUBMISSION
cursor.execute(f'''CREATE TABLE Sobreviventes (PassengerId INT, Survived INT)''')

conex.commit()
```

```python
# Inserção dos dados na tabela gender_submission
for row in df.itertuples():
    cursor.execute('''
                INSERT INTO Sobreviventes (PassengerId, Survived)
                VALUES (?,?)
                ''',
                row.PassengerId,
                row.Survived
                )

conex.commit()
```
